# MoeaBench Example 04: Statistical Repeats

Running an experiment multiple times to analyze statistical stability.

```python
# Install MoeaBench
!pip install --quiet git+https://github.com/opensciware/moeabench.git
```

```python
import MoeaBench as mb

exp = mb.experiment()
exp.benchmark = mb.benchmarks.DTLZ2()
exp.moea      = mb.moeas.NSGA3(population=50, generations=50)

# Run 5 times (seeds will auto-increment)
exp.run(5)

# Calculate Hypervolume Matrix (Generations x Runs)
hv = mb.hv(exp)

# Plot Mean and Standard Deviation Shadow
mb.timeplot(hv, show_bounds=True)
```

```python
# Plot all 5 fronts combined (Superfront)
mb.spaceplot(exp.superfront(), title="Combined Non-Dominated Front")
```

```python
# Plot each run individually to check spread
mb.spaceplot(*exp.all_fronts(), title="All 5 Runs")
```
