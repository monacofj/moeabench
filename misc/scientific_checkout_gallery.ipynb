# MoeaBench v0.7.6: Scientific Checkout Gallery
This notebook provides a step-by-step visual verification of the **Scientific Checkout (Phase 1.5)**.

For each MOP and Algorithm, we plot the **Analytical Optimal (v0.7.6 Ground Truth)** against the **Approximated Front** generated in real-time.

```python
import os
import sys
import matplotlib.pyplot as plt

# Discover Root
curr = os.getcwd()
for _ in range(3):
    if os.path.exists(os.path.join(curr, "MoeaBench")):
        PROJECT_ROOT = curr
        break
    curr = os.path.dirname(curr)
else:
    PROJECT_ROOT = os.getcwd()

if PROJECT_ROOT not in sys.path:
    sys.path.append(PROJECT_ROOT)

import MoeaBench as mb
from MoeaBench.core import SmartArray
print(f"Project Root: {PROJECT_ROOT}")
```

## 1. DTLZ2: The Spherical Baseline

```python
mop = mb.mops.DTLZ2(M=3)
exp = mb.experiment(mop=mop)
F_opt = exp.optimal(n_points=1000).objectives

exp.run(mb.moeas.NSGA2(), n_gen=200, pop_size=100)
F_res = exp.pop().objectives

mb.view.topo_shape(
    SmartArray(F_opt, name="Analytical Optimal (GT)"),
    SmartArray(F_res, name="NSGA-II Result"),
    title="DTLZ2 (M=3) Checkout"
)
```
