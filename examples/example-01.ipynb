# MoeaBench Example 01: Simple Run

This example demonstrates how to set up a basic experiment with `DTLZ2` and `NSGA3`, run it, and visualize the results.

```python
# Install MoeaBench
!pip install --quiet git+https://github.com/opensciware/moeabench.git
```

```python
import mb_path
import MoeaBench as mb

# 1) Create and configure experiment
exp = mb.experiment()
exp.mop = mb.mops.DTLZ2()
exp.moea      = mb.moeas.NSGA3(population=50, generations=50)

# 2) Run the experiment
exp.run()

# 3) Visualize results
# Plot the Pareto Front
mb.spaceplot(exp.front())
```

```python
# Calculate and Plot Hypervolume
hv = mb.hv(exp)
mb.timeplot(hv)
```
