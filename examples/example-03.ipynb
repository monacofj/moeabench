# MoeaBench Example 03: Trajectory Analysis

Analysis of populations at different stages (Early vs Late).

```python
# Install MoeaBench
!pip install --quiet git+https://github.com/opensciware/moeabench.git
```

```python
import MoeaBench as mb

# Setup Experiment
exp = mb.experiment()
exp.benchmark = mb.benchmarks.DTLZ2()
exp.moea      = mb.moeas.NSGA3(population=100, generations=50)
exp.run()

# 1) Plot Dominated (Blue) vs Non-Dominated (Orange)
mb.spaceplot(exp.non_front(), exp.front(), title="Dominated vs Non-Dominated")
```

```python
# 2) Compare Early Stage (Gen 5) vs Final Stage (Gen 50)
mb.spaceplot(exp.front(5), exp.front(), title="Early vs Final")
```
