# Example 17: Convergence, Proximity, and Distributional Correlation Analysis


This example explores the different ways to measure "Quality" in MOAs using distance-based and distribution-based metrics.

```python
# Install MoeaBench
!pip install --quiet git+https://github.com/monacofj/moeabench.git
```

### Setup


```python
from MoeaBench import mb
mop = mb.mops.DTLZ2(M=2)
exp = mb.experiment()
exp.name = "Metric Analysis Study"
exp.mop = mop
exp.moea = mb.moeas.NSGA2(population=100, generations=50)
```

### Execution


```python
exp.run(repeat=1)
```

### Calculating metrics


```python
res_igd = mb.metrics.igd(exp)
res_gd = mb.metrics.gd(exp)
res_emd = mb.metrics.emd(exp)
```

### Correlation Analysis


```python
mb.metrics.plot_matrix(res_igd, res_gd, res_emd, title="Metric Correlation Portfolio")
```

### Comparing Trajectories


```python
mb.view.perf_history(res_igd, res_gd, res_emd, title="Traj: Distances vs Distribution")
```

### Interpretation

IGD, GD, and EMD provide different perspectives on convergence and topological integrity.
