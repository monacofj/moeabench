# Simple Experiment with MoeaBench

This notebook demonstrates how to run a simple experiment using MoeaBench with Benchmark DTLZ2 and MOEA NSGA-II.

```python
from MoeaBench import mb

exp = mb.experiment()

exp.name = 'My Experiment'

exp.benchmark = mb.benchmarks.DPF5()  # Choose the problem
exp.moea      = mb.moeas.NSGA3()      # and the algorithm.

exp.run()

print(exp.objectives)
```
